```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes= ["Python","R","Ai","Ml","Ds"]
class1_student = [23,46,35,20,40]
class2_student = [20,40,50,60,10] 
class3_student = [50,25,45,35,25]
```

```python
style.use("ggplot") #only run one time and use mutipal time
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.3,align='edge',color="y",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
plt.bar(classes,class2_student,width=0.3,align='edge',color="g",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
plt.bar(classes,class3_student,width=0.3,align='edge',color="b",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False

#plt.show()
plt.legend()
```

```python
plt.barh(classes,class1_student) #show horizontal use "plt.barh" 
```

```python
plt.figure(figsize=(16,9)) 
plt.barh(classes,class1_student,align='edge',color="y",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
#plt.show()
plt.legend()
```

# Tital or label use in bar chat

```python
style.use("ggplot") #only run one time and use mutipal time
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.3,align='edge',color="y",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
plt.title("Bar chat for Student",fontsize=15)
plt.xlabel("classes")
plt.ylabel("No. of student",fontsize=20)
#plt.show()
plt.legend()
```

below code show how we can show diff diff "bar chat" using "class_index" & "width"
{classes_index = (np.arange(len(classes))) store total number value
width =0.3} show total size that can get
1,...> [plt.bar(classes_index,class1_student)]
2,...> [plt.bar(classes_index+width,class2_student)]
3,...>[plt.bar(classes_index+width+width,class3_student)]
 
 key ponit


```python
style.use("ggplot") #only run one time and use mutipal time
plt.figure(figsize=(16,9))
classes_index = (np.arange(len(classes)))
width =0.3
plt.bar(classes_index,class1_student,width,align='edge',color="y",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
plt.bar(classes_index+width,class2_student,width,align='edge',color="g",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class2_student",) #visible=False
plt.bar(classes_index+width+width,class3_student,width,align='edge',color="b",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class3_student",) #visible=False
plt.xticks(classes_index+width,classes,rotation=20)
plt.title("Bar chat Graph for student",fontsize=25)
plt.xlabel("classes", fontsize=25)
plt.ylabel("No. of Student",fontsize=25)
#plt.show()
plt.legend()
```

```python
                          #show horizontal use "barh % ytrick"

style.use("ggplot") #only run one time and use mutipal time
plt.figure(figsize=(16,9))
classes_index = (np.arange(len(classes)))
width =0.3
plt.barh(classes_index,class1_student,width,align='edge',color="y",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class1_student",) #visible=False
plt.barh(classes_index+width,class2_student,width,align='edge',color="g",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class2_student",) #visible=False
plt.barh(classes_index+width+width,class3_student,width,align='edge',color="b",edgecolor ="r",linewidth=5,alpha =0.8,linestyle="--",label="class3_student",) #visible=False
plt.yticks(classes_index+width,classes,rotation=20)
plt.title("Bar chat Graph for student",fontsize=25)
plt.xlabel("classes", fontsize=25)
plt.ylabel("No. of Student",fontsize=25)
#plt.show()
plt.legend()
```
